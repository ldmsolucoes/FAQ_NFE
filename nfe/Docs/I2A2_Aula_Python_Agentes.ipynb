## Listas, dicionrios, tuplas, sets - Compreenses de listas e dicionrios


* **Listas - "list" - []**: Vetores ou Arrays. So colees mutveis e ordenadas de itens. Suportam operaes `append()`, `extend()`, `insert()`, `remove()`, `pop()`, e atribuies de ndice (`list[index] = value`).


````python
# Criando uma lista de nmeros inteiros
minha_lista_numeros:list[int] = [10, 20, 30, 40, 20]
print(f"Lista inicial: {minha_lista_numeros}")

# Acessando elementos por ndice (lembre-se: o primeiro elemento  o ndice 0)
print(f"Primeiro elemento: {minha_lista_numeros[0]}")
print(f"Terceiro elemento: {minha_lista_numeros[2]}")

# Modificando um elemento
minha_lista_numeros[1] = 25
print(f"Lista aps modificar o segundo elemento: {minha_lista_numeros}")

# Adicionando um elemento ao final da lista
minha_lista_numeros.append(50)
print(f"Lista aps adicionar '50': {minha_lista_numeros}")

# Removendo um elemento pelo valor
minha_lista_numeros.remove(20) # Remove a primeira ocorrncia de '20'
print(f"Lista aps remover a primeira ocorrncia de '20': {minha_lista_numeros}")

# Removendo um elemento pela posio (ndice)
elemento_removido = minha_lista_numeros.pop(3) # Remove o elemento no ndice 3 (40)
print(f"Lista aps remover elemento no ndice 3: {minha_lista_numeros}")
print(f"Elemento removido com pop(): {elemento_removido}")

# Retornando partes da lista - [10, 25, 30, 50]
print(f"Minha lista: {minha_lista_numeros}")
print(f"Retornando o tamanho da lista: {len(minha_lista_numeros)}")
print(f"Retornando o ltimo elemento: {minha_lista_numeros[-1]}")
print(f"Retornando o 2 e 3 elementos: {minha_lista_numeros[1:3]}")
print(f"Retornando do 2 elemento ao final da lista: {minha_lista_numeros[1:]}")
print(f"Retornando at o 3 elemento da lista: {minha_lista_numeros[:3]}")

# Iterando sobre uma lista
print("Iterando sobre a lista:")
for numero in minha_lista_numeros:
    print("\t-", numero)

# Verificando se um elemento existe na lista
if 25 in minha_lista_numeros:
    print("25 est na lista.")
````


* **Tuplas - "tuple" - ()**: Vetores ou Arrays. So colees **imutveis** e ordenadas de itens. Uma vez criada, os elementos de uma tupla no podem ser alterados, adicionados ou removidos.

````python
from typing import Any

# Criando uma tupla de informaes de um usurio (nome, idade, cidade)
dados_usuario:tuple[Any] = ("Alice", 30, "Porto Alegre")
print(f"Tupla inicial: {dados_usuario}")

# Acessando elementos por ndice
print(f"Nome do usurio: {dados_usuario[0]}")
print(f"Idade do usurio: {dados_usuario[1]}")

# Tentando modificar um elemento (isso causaria um TypeError)
try:
    dados_usuario[0] = "Maria"
except TypeError as e:
    print(f"Erro ao tentar modificar uma tupla: {e}")
except Exception as e:
    print(f"Erro: {e}")
finally:
    pass

# Tuplas podem ser usadas como chaves em dicionrios (seus elementos so hashable)
# Exemplo: um dicionrio que armazena a localizao de cidades
locais = {
    ("So Paulo", -23.55, -46.63): "Capital de SP",
    ("Rio de Janeiro", -22.90, -43.17): "Cidade Maravilhosa"
}
print(f"Localizaes de cidades: {locais}")
print(f"Descrio de So Paulo: {locais[('So Paulo', -23.55, -46.63)]}")
````

* **Dicionrios - "dict" - {"key": "value"}**: Mapeamento Chave-Valor, cada item  um par chave: valor.

````python
# Criando um dicionrio de informaes de um produto
produto:dict[str, Any] = {
    "id": "A123",
    "nome": "Smartphone X",
    "preco": 1200.50,
    "estoque": 50,
    "cores_disponiveis": ["preto", "branco", "azul"]
}
print(f"Dicionrio inicial do produto: {produto}")

# Acessando valores usando as chaves
print(f"Nome do produto: {produto['nome']}")
print(f"Preo do produto: R${produto['preco']}")

try:
    print(f"Ean do produto: {produto['ean']}")
except KeyError as e:
    print(f"Erro: {e} no existe no dicionrio.")

# Acessando valores usando o get
print(f"Nome do produto (get): {produto.get('nome', None)}")
print(f"Ean do produto no existe: {produto.get('ean', None)}")

# Modificando um valor
produto["preco"] = 1150.00
print(f"Novo preo do produto: R${produto['preco']}")

# Adicionando um novo par chave-valor
produto["fabricante"] = "TechCorp"
print(f"Dicionrio aps adicionar fabricante: {produto}")

# Removendo um par chave-valor
estoque_removido:Any = produto.pop("estoque")
print(f"Dicionrio aps remover estoque: {produto}")
print(f"Estoque removido: {estoque_removido}")

# Iterando sobre as chaves, valores ou itens (pares chave-valor)
print("\nChaves do dicionrio:")
for chave in produto.keys():
    print(chave)

print("\nValores do dicionrio:")
for valor in produto.values():
    print(valor)

print("\nItens (chave-valor) do dicionrio:")
for chave, valor in produto.items():
    print(f"{chave}: {valor}")

# Verificando se uma chave existe
if "nome" in produto:
    print("A chave 'nome' existe no dicionrio.")
````

* **Sets - "set" - {"value"}**: Todos os elementos devem ser nicos. Itens duplicados so automaticamente descartados na insero.

````python
# Criando um set de nmeros (duplicatas so automaticamente ignoradas)
meu_set_numeros:set[int] = {1, 2, 3, 4, 2, 5}
print(f"Set inicial (observe que o '2' duplicado foi removido): {meu_set_numeros}")

# Adicionando um elemento
meu_set_numeros.add(6)
print(f"Set aps adicionar '6': {meu_set_numeros}")

# Tentando adicionar um elemento que j existe (no far diferena)
meu_set_numeros.add(3)
print(f"Set aps tentar adicionar '3' novamente: {meu_set_numeros}")

# Removendo um elemento
meu_set_numeros.remove(1) # Lana um KeyError se o elemento no existir
print(f"Set aps remover '1': {meu_set_numeros}")

# Usando discard() para remover (no lana erro se o elemento no existir)
meu_set_numeros.discard(10) # 10 no est no set, mas no causa erro
print(f"Set aps tentar descartar '10': {meu_set_numeros}")

# Operaes de conjunto
set_a:set[int] = {1, 2, 3, 4}
set_b:set[int] = {3, 4, 5, 6}

print(f"\nSet A: {set_a}")
print(f"Set B: {set_b}")

# Unio (elementos em A OU B)
uniao = set_a.union(set_b)
print(f"Unio de A e B: {uniao}") # {1, 2, 3, 4, 5, 6}

# Interseo (elementos em A E B)
intersecao:set[int] = set_a.intersection(set_b)
print(f"Interseo de A e B: {intersecao}") # {3, 4}

# Diferena (elementos em A mas NO em B)
diferenca_ab:set[int] = set_a.difference(set_b)
print(f"Diferena A - B: {diferenca_ab}") # {1, 2}

# Diferena simtrica (elementos em A OU B, mas NO em ambos)
diferenca_simetrica:set[int] = set_a.symmetric_difference(set_b)
print(f"Diferena simtrica A ^ B: {diferenca_simetrica}") # {1, 2, 5, 6}

# Verificando se um elemento existe no set
if 5 in meu_set_numeros:
    print("5 est no set.")
````

## Funes e parmetros

````python
def soma(a:int, b:int, c:int=1) -> int:
    # Docstring - Boas prticas
    """
    Funo que retorna a soma de dois nmeros inteiros.
    Args:
        a (int): Primeiro nmero.
        b (int): Segundo nmero.
        c (int, optional): Multiplicador. Padro  1.
    Returns:
        int: Soma de A + B multiplicados por C.
    """
    return (a + b) * c


# Chamando a funo com vrios parmetros
resultado = soma(10, 5)
print(f"Resultado da soma(10, 5): {resultado}")

resultado = soma(10, 5, 2)
print(f"Resultado da soma(10, 5, 2): {resultado}")

resultado = soma(a=10, b=5)
print(f"Resultado da soma(a=10, b=5): {resultado}")

resultado = soma(10, 5, c=2)
print(f"Resultado da soma(10, 5, c=2): {resultado}")

resultado = soma(b=5, c=2, a=10)
print(f"Resultado da soma(b=5, c=2, a=10): {resultado}")
````

## Escopo de variveis

1. **Escopo Local**: Uma varivel com escopo local  criada dentro de uma funo. Ela s existe enquanto a funo est sendo executada e s pode ser acessada de dentro daquela funo. Assim que a funo termina, a varivel local  destruda.

````python
def minha_funcao() -> None:
    # 'x'  uma varivel local, criada dentro de minha_funcao
    x:int = 10
    print(f"Dentro da funo, x vale: {x}")

# Chamando a funo
minha_funcao()

# Tentando acessar 'x' fora da funo resultar em um erro
try:
    print(x)
except NameError as e:
    print(f"Erro ao tentar acessar 'x' fora da funo: {e}")
````

2. **Escopo Global**: Uma varivel com escopo global  criada fora de qualquer funo ou classe, diretamente no nvel superior do seu script. Ela pode ser acessada de qualquer lugar do cdigo, tanto dentro quanto fora de funes.

````python
# 'y'  uma varivel global, criada no nvel superior do script
y:int = 20

def outra_funcao() -> None:
    print(f"Dentro da funo, posso acessar 'y' global: {y}")

outra_funcao()
print(f"Fora da funo, posso acessar 'y' global: {y}")
````

**Cuidado com a modificao de variveis globais dentro de funes**:
Por padro, se voc tenta atribuir um novo valor a uma varivel dentro de uma funo, o Python cria uma nova varivel local com o mesmo nome, em vez de modificar a global.

````python
z:int = 30 # Varivel global

def tenta_modificar_z():
    z = 40 # Isso cria uma NOVA varivel LOCAL 'z'
    print(f"Dentro da funo, z local vale: {z}")

tenta_modificar_z()
print(f"Fora da funo, z global ainda vale: {z}") # O 'z' global no foi alterado!
````

Para realmente modificar uma varivel **global** dentro de uma funo, voc precisa usar a palavra-chave `global`. Use global com moderao, pois o uso excessivo pode tornar o cdigo mais difcil de entender e manter.

````python
w:int = 50 # Varivel global

def modifica_w_global() -> None:
    global w # Declara que estamos nos referindo  varivel 'w' global
    w = 60 # Agora, isso modifica a varivel global 'w'
    print(f"Dentro da funo (global), w vale: {w}")

modifica_w_global()
print(f"Fora da funo, w global agora vale: {w}") # O 'w' global foi alterado!
````

3. **Escopo Enclosing (Nonlocal)**: Este escopo  um pouco mais especfico e se aplica a funes aninhadas (uma funo definida dentro de outra). Uma varivel no escopo "enclosing" (tambm conhecido como "no local")  acessvel pela funo interna, mas no  global.

````python
def funcao_externa() -> None:
    mensagem:str = "Ol do escopo enclosing!" # Varivel no escopo enclosing

    def funcao_interna() -> None:
        print(f"Dentro da funo interna: {mensagem}") # Acessando 'mensagem' do escopo enclosing
    funcao_interna()

funcao_externa()

# Tentando acessar 'mensagem' fora da funcao_externa (resulta em erro)
try:
    print(mensagem)
except NameError as e:
    print(f"Erro ao tentar acessar 'mensagem' fora do escopo enclosing: {e}")
````

Assim como no global, se voc quiser modificar uma varivel do escopo **"enclosing"** dentro da funo aninhada, precisa usar a palavra-chave `nonlocal`

````python
def contador() -> int:
    num:int = 0 # Varivel no escopo enclosing

    def incrementar() -> None:
        nonlocal num # Declara que estamos nos referindo  'num' do escopo enclosing
        num += 1
        print(f"Nmero incrementado para: {num}")

    return incrementar # Retorna a funo interna

meu_contador = contador() # 'meu_contador' agora  a 'incrementar'
meu_contador()
meu_contador()
````

## Atributos e mtodos (pblicos, protegidos (`_`) e privados (`__`))

Diferente de algumas linguagens como Java ou C++, **Python** no tem uma imposio estrita de visibilidade via modificadores de acesso (como `public`, `protected`, `private`). Em vez disso, Python usa uma conveno de nomenclatura e um mecanismo conhecido como **"name mangling"** para sugerir ou controlar o acesso.

1. **Atributo/Mtodo Pblico (Public)**
* **Conveno**: No h prefixo especial.
* **Acesso**: Pode ser acessado e modificado livremente de qualquer lugar (dentro ou fora da classe).

2. **Atributo/Mtodo Protegido (Protected)**
* **Conveno**: O nome do atributo comea com um nico underscore _ (ex: _atributo).
* **Acesso**: Em Python, um atributo protegido ainda pode ser acessado de fora da classe, mas a conveno _ serve como um `aviso` para outros desenvolvedores: "*Este atributo  destinado a uso interno da classe ou de suas subclasses, e no deve ser acessado ou modificado diretamente de fora.*"
3. **Atributo/Mtodo Privado (Private)**
* **Conveno**: O nome do atributo comea com dois underscores __ (ex: __atributo).
* **Acesso**: Atributo/Mtodo privados passam por um processo chamado **"name mangling"** (embaralhamento de nomes). O interpretador Python renomeia o atributo internamente para `_NomeDaClasse__nome_do_atributo_ou_metodo`. Isso torna o acesso direto de fora da classe **muito difcil**, mas no `impossvel`. O acesso  quase restrito  prpria classe.

````python
class Produto:
    def __init__(self, nome:str, preco:float) -> None:
        self.nome = nome     # pblico
        self._preco = preco  # protegido
        self.__estoque = 100 # privado

    # pblico
    def exibir(self) -> str:
        return f"Produto: {self.nome}, Preo: {self._preco:.2f}, Estoque: {self.__estoque}"

    # privado
    def __atualizar_estoque(self, quantidade:int) -> None:
        if self.__estoque + quantidade < 0:
            raise ValueError("Quantidade insuficiente no estoque.")
        self.__estoque += quantidade

    # pblico
    def adicionar_estoque(self, quantidade:int) -> None:
        self.__atualizar_estoque(quantidade)

    # protegido
    def _remover_estoque(self, quantidade:int) -> None:
        self.__atualizar_estoque(quantidade*-1)

    # pblico
    def estoque(self) -> int:
        return self.__estoque
````

````python
# Instncia da classe de Produto
produto_teclado:Produto = Produto(nome="Teclado", preco=129.90)

# Modificando atributo pblico
produto_teclado.nome = "Teclado Mecnico"

# Modificando atributo protegido (NO  RECOMENDADO)
produto_teclado._preco = 119.90

print(produto_teclado.exibir())
print(f"Estoque atual: {produto_teclado.estoque()}")

# NO  RECOMENDADO, mas no causa erro.
produto_teclado.__estoque = 90  # Isso no funcionar, pois __estoque  privado
print(f"Tentativa em alterar o estoque: {produto_teclado.__estoque} / estoque atual: {produto_teclado.estoque()}")

produto_teclado.endereco = "Rua Exemplo, 123"  # Atributo dinmico (no recomendado)
print(f"Tentativa em adicionar um campo : {produto_teclado.endereco}")
````

## Mtodos de instncia, \*args e \*\*kwargs

Um mtodo de instncia  o tipo de mtodo mais comum em uma classe Python. Ele opera sobre uma instncia especfica (ou seja, um objeto) da classe. Isso significa que ele tem acesso aos atributos e a outros mtodos dessa instncia.

**Caractersticas Principais:**

1. Primeiro Parmetro `self`: O primeiro parmetro de um mtodo de instncia deve ser sempre `self` (por conveno, embora voc possa usar qualquer nome). `self`  uma referncia  prpria instncia do objeto que est chamando o mtodo.  atravs de `self` que voc acessa os atributos (self.atributo) e outros mtodos (self.metodo()) daquela instncia.

2. Operam sobre o **Estado do Objeto**: Eles so projetados para interagir com os dados (atributos) que pertencem a uma instncia especfica da classe.

3. Precisam de uma Instncia: Voc precisa criar um objeto (uma instncia) da classe para poder chamar um mtodo de instncia.

**Variveis args e kwargs**

|                    | `*args`                                                      | `**kwargs`                                             |
| ------------------ | ------------------------------------------------------------ | ------------------------------------------------------ |
| Descrio             | Recebe vrios argumentos posicionais (sem nome)  Vira uma tupla |  Recebe vrios argumentos nomeados (chave=valor)  Vira um dicionrio|
| Coleta             | Argumentos posicionais (no nomeados)                        | Argumentos nomeados (chave=valor)                      |
| Tipo               | Tupla                                                        | Dicionrio                                             |
| Quando usar        | Quando voc no sabe quantos argumentos o usurio vai passar | Quando voc quer aceitar argumentos com nomes variados |
| Exemplo de chamada | `func(1, 2, 3)`                                              | `func(nome='Ana', idade=25)`                           |

O nome (args, kwargs)  s uma conveno. O que importa de verdade so os \* e \*\*.

````python
class Pedido:
    def __init__(self, id_pedido:str) -> None:
        self.id_pedido:str = id_pedido
        self.itens:list = []
        self.observacoes:dict = {}
        print(f"Pedido {self.id_pedido} criado.")

    def adicionar_itens(self, nome_cliente:str, *args, **kwargs) -> None:
        """
        Adiciona itens e detalhes a um pedido especfico.

        Parmetros:
        - nome_cliente (str): O nome do cliente que fez o pedido (parmetro normal).
        - *args (args): Argumentos posicionais variveis, representando os itens do pedido.
                         Cada 'item' pode ser uma string como "Camiseta P".
        - **kwargs (kwargs): Argumentos de palavra-chave variveis,
                                         para notas ou informaes adicionais do pedido.
                                         Ex: frete='expresso', embalagem='presente'.
        """
        print(f"\n--- Adicionando itens ao Pedido {self.id_pedido} para o cliente {nome_cliente} ---")
        print("args:", args)  # Exibe os itens recebidos como argumentos posicionais
        print("kwargs:", kwargs, '\n\n')

        # Para declararmos o args e o kwargs, usamos o * e ** respectivamente
        # Para passarmos para um outro mtodo tambm usamos essas notaes
        # Exemplo: chamando_outro_metodo(*arg, **kwargs)

        # Adicionando itens (vindos de *itens)
        if args: # Verifica se a tupla 'itens' no est vazia
            for item in args:
                self.itens.append(item)
                print(f"  - Item adicionado: {item}")
        else:
            print("  Nenhum item principal fornecido.")

        # Adicionando detalhes adicionais (vindos de **detalhes_adicionais)
        if kwargs: # Verifica se o dicionrio 'detalhes_adicionais' no est vazio
            print("  Detalhes adicionais:")
            for chave, valor in kwargs.items():
                self.observacoes[chave] = valor
                print(f"    - {chave.replace('_', ' ').capitalize()}: {valor}")
        else:
            print("  Nenhum detalhe adicional fornecido.")

        print(f"Status atual do Pedido {self.id_pedido}:")
        print(f"  Itens: {self.itens if self.itens else 'Nenhum'}")
        print(f"  Observaes: {self.observacoes if self.observacoes else 'Nenhum'}")
````

## Herana simples e mltipla

**Herana**  um dos pilares da Programao Orientada a Objetos, permitindo que voc crie novas classes que herdam caractersticas (atributos e mtodos) de classes existentes. Isso promove a reutilizao de cdigo e a organizao hierrquica.

1. **Herana Simples**: Na herana simples, uma classe (chamada classe filha ou subclasse) herda de apenas uma outra classe (chamada classe pai ou superclasse).  o tipo de herana mais comum e direto.

**Exemplo**: ` um(a)`  Um Cachorro ** um** Animal.

Herana Simples

````python
class Animal: # class Animal(object):
    def __init__(self, nome:str, especie:str) -> None:
        self.nome = nome
        self.especie = especie
        print(f"Um novo animal chamado {self.nome} ({self.especie}) foi criado.")

    def comer(self) -> None:
        print(f"{self.nome} est comendo.")

    def dormir(self) -> None:
        print(f"{self.nome} est dormindo.")
````

````python
class Cachorro(Animal):
    def __init__(self, nome:str, raca:str) -> None:
        # Chama o construtor da classe pai (Animal)
        super().__init__(nome=nome, especie="Canino")
        self.raca = raca
        print(f"  -->  um cachorro da raa {self.raca}.")

    # Mtodo especfico de Cachorro
    def latir(self) -> None:
        print(f"{self.nome} est latindo: Au au!")

    # Sobrescrevendo o mtodo comer() da superclasse
    def comer(self) -> None:
        print(f"{self.nome} est comendo rao para cachorro.")
````

````python
class Gato(Animal):
    def __init__(self, nome:str, cor_pelo:str) -> None:
        super().__init__(nome, especie="Felino")
        self.cor_pelo = cor_pelo
        print(f"  -->  um gato de pelo {self.cor_pelo}.")

    # Mtodo especfico de Gato
    def miar(self) -> None:
        print(f"{self.nome} est miando: Miau!")
````

````python
# Criando instncias
rex = Cachorro(nome="Rex", raca="Pastor Alemo")
rex.comer()   # Chama o mtodo 'comer' sobrescrito em Cachorro
rex.latir()   # Chama o mtodo especfico de Cachorro
rex.dormir()  # Chama o mtodo 'dormir' herdado de Animal

print('', "-" * 20, '', sep='\n')

tom = Gato(nome="Tom", cor_pelo="Cinza")
tom.comer()  # Chama o mtodo 'comer' herdado de Animal (no sobrescrito em Gato)
tom.miar()   # Chama o mtodo especfico de Gato
````

2. **Herana Mltipla**: Na herana mltipla, uma classe (subclasse) herda de duas ou mais classes (superclasses). Isso permite combinar funcionalidades de diferentes fontes em uma nica classe.

**Exemplo**: ` um(a) e tambm  um(a)`  Um Anfibio  um `AnimalAquatico` e tambm  um `AnimalTerrestre`.

**Pontos Chave e Desafios:**

* `MRO` **(Method Resolution Order)**: Python usa um algoritmo chamado `C3 Linearization` para determinar a ordem em que as superclasses so pesquisadas por mtodos e atributos. Voc pode ver a MRO de uma classe usando MinhaClasse.mro() ou help(MinhaClasse). Isso  crucial para resolver conflitos de nomes (o "problema do diamante").

* `super()` com **Herana Mltipla**: O uso de `super()` em herana mltipla  a forma recomendada para chamar mtodos da superclasse, pois ele respeita a MRO e garante que cada mtodo na cadeia de herana seja chamado uma nica vez e na ordem correta, mesmo em estruturas complexas.

* **Problema do Diamante**: Ocorre quando uma classe herda de duas classes que, por sua vez, herdam de uma classe comum, formando um "diamante" na hierarquia. A MRO de Python lida bem com isso, garantindo que o mtodo da classe comum seja executado apenas uma vez.

Herana Multipla

````python
class Voador:
    def __init__(self, altura_maxima:int) -> None:
        print("Criando uma entidade Voadora.")
        self.altitude_maxima = altura_maxima

    def voar(self) -> None:
        print("Voando alto no cu!")
````

````python
class Nadador:
    def __init__(self, profundidade_maxima:int) -> None:
        print("Criando uma entidade Nadadora.")
        self.profundidade_maxima = profundidade_maxima

    def nadar(self) -> None:
        print("Nadando nas profundezas!")
````

````python
# Pato herda de Voador E Nadador
class Pato(Voador, Nadador):
    def __init__(self, nome:str) -> None:
        print(f"Criando um Pato chamado {nome}.")
        self.nome = nome
        # Chama os construtores das classes base respeitando a MRO
        # Isso  fundamental para herana mltipla com construtores
        Voador.__init__(self, altura_maxima=1000) # Chama o __init__ de Voador
        Nadador.__init__(self, profundidade_maxima=5) # Chama o __init__ de Nadador

    def grasnar(self) -> None:
        print(f"{self.nome} est grasnando: Quack! Quack!")
````

````python
# Criando uma instncia
donald = Pato(nome="Donald")

# Pato tem mtodos de Voador
donald.voar()
print(f"Altitude mxima do {donald.nome}: {donald.altitude_maxima}m")

# Pato tem mtodos de Nadador
donald.nadar()
print(f"Profundidade mxima do {donald.nome}: {donald.profundidade_maxima}m")

# Pato tem seus prprios mtodos
donald.grasnar()

# Verificando a MRO (Method Resolution Order)
print("\nMRO do Pato:")
print(Pato.mro())

# Sada esperada da MRO:
# [<class '__main__.Pato'>, <class '__main__.Voador'>, <class '__main__.Nadador'>, <class 'object'>]
# Isso significa que, se um mtodo no for encontrado em Pato,
# ele procurar em Voador,
# depois em Nadador, e por ltimo em 'object' (a classe base de todas as classes Python).
````

## Classes abstratas (`ABC`, `@abstractmethod`)

Em Programao Orientada a Objetos, uma **classe abstrata**  uma classe que **no pode ser instanciada** diretamente. Seu principal propsito  definir uma interface ou um contrato que suas subclasses devem seguir. Elas atuam como um template ou um esquema para outras classes.

* Classes abstratas podem ter:

1. **Mtodos concretos**: Mtodos com implementao padro.
2. **Mtodos abstratos**: Mtodos que so declarados, mas no implementados. A implementao  deixada para as subclasses.

Em Python, voc consegue criar classes abstratas usando o mdulo `abc` (Abstract Base Classes) da biblioteca padro, em conjunto com o decorador `@abstractmethod`.

**Por que usar Classes Abstratas?**

1. **Definir um Contrato (Interface)**: Garantem que qualquer subclasse implemente certos mtodos essenciais. Isso impe uma estrutura e evita que desenvolvedores se esqueam de funcionalidades crticas.

2. **Reuso de Cdigo**: Voc pode colocar mtodos concretos e atributos comuns a todas as subclasses na classe abstrata.

3. **Polimorfismo**: Permitem que voc trate diferentes tipos de objetos de maneira uniforme, desde que eles compartilhem a mesma interface definida pela classe abstrata.

**pass** vs **... (Ellipsis)**

|                       | `pass`                                                     | `...` (Ellipsis)                                                                                         |
| --------------------- | ---------------------------------------------------------- | -------------------------------------------------------------------------------------------------------- |
| **O que ?**          | Uma **instruo nula**                                     | Um **objeto especial** chamado `Ellipsis`                                                                |
| **Funo principal**  | Serve para deixar o bloco vazio, **sem erro de sintaxe**   | Tambm pode ser usado como placeholder, mas  um **objeto de verdade** (no uma palavra-chave)           |
| **Erro se no usar?** | Sim! Se o bloco estiver vazio, precisa de algo como `pass` | No d erro se voc colocar `...`, mas o comportamento  diferente                                       |
| **Tipo**              | Palavra-chave da linguagem                                 | Literal que representa o tipo `Ellipsis`                                                                 |
| **Uso mais comum**    | Bloco de cdigo vazio (em funes, classes, ifs, etc)      | Placeholder para indicar "aqui vai ter cdigo no futuro", usado tambm em **fatiamento de arrays NumPy** |


````python
from abc import ABC, abstractmethod

class MetodoPagamento(ABC):
    def __init__(self, valor:float) -> None:
        self.valor:float = valor
        self.status:str = "pendente"
        print(f"Mtodo de pagamento para R${valor:.2f} criado. Status: {self.status}")

    # Este mtodo DEVE ser implementado pelas subclasses
    @abstractmethod
    def processar_pagamento(self) -> bool:
        """Processa o pagamento de acordo com o mtodo especfico."""
        pass # Nenhuma implementao aqui

    # Este mtodo DEVE ser implementado pelas subclasses
    @abstractmethod
    def verificar_status(self) -> str:
        """Verifica o status atual do pagamento."""
        ... # Nenhuma implementao aqui,  apenas um placeholder

    # Mtodo concreto (com implementao)
    def exibir_detalhes_comuns(self):
        """Exibe detalhes comuns a todos os mtodos de pagamento."""
        print(f"Valor: R${self.valor:.2f}, Status Atual: {self.status}")
````

````python
class CartaoCredito(MetodoPagamento):
    def __init__(self, valor:float, numero_cartao:str, cvc:str) -> None:
        # Chama o construtor da classe abstrata
        super().__init__(valor)
        self.numero_cartao = numero_cartao
        self.cvc = cvc
        print("  --> Tipo: Carto de Crdito")

    # Implementao OBRIGATRIA
    def processar_pagamento(self) -> bool:
        print(f"Processando R${self.valor:.2f} via Carto de Crdito (final {self.numero_cartao[-4:]})...")
        self.status = "aprovado" if self.valor < 1000 else "recusado" # Exemplo de lgica
        print(f"Pagamento via Carto de Crdito {self.status}.")
        return self.status == "aprovado"

    # Implementao OBRIGATRIA
    def verificar_status(self) -> str:
        print(f"Verificando status do pagamento de Carto de Crdito: {self.status}.")
        return self.status
````

````python
class Pix(MetodoPagamento):
    def __init__(self, valor:float, chave_pix:str) -> None:
        super().__init__(valor)
        self.chave_pix = chave_pix
        self.codigo_qr_gerado:str = "ABC123XYZ"
        print("  --> Tipo: Pix")

    # Implementao OBRIGATRIA
    def processar_pagamento(self) -> bool:
        print(f"Gerando QR Code Pix para R${self.valor:.2f} (Chave: {self.chave_pix})...")
        print(f"Use o cdigo QR: {self.codigo_qr_gerado} para pagar.")
        self.status = "aguardando_pagamento"
        # Em um sistema real, haveria um webhook para mudar o status
        return True

    # Implementao OBRIGATRIA
    def verificar_status(self) -> str:
        if self.status == "aguardando_pagamento" and self.valor > 500:
             self.status = "aprovado"
        print(f"Verificando status do pagamento Pix: {self.status}.")
        return self.status
````

## Mtodos de classe (`@classmethod`)

Os **mtodos de classe** em Python so um tipo de mtodo que opera sobre a classe em si, e no sobre uma instncia especfica da classe. Eles recebem a prpria classe como primeiro argumento (por conveno, chamado `cls`), em vez da instncia (`self`).

Voc define um mtodo de classe usando o decorador `@classmethod` logo acima da sua definio de mtodo.



````python
import json
from datetime import date

class Pessoa:

    def __init__(self, nome_completo:str, idade:int) -> None:
        self.nome_completo = nome_completo
        self.idade = idade
        print(f"Pessoa '{self.nome_completo}' ({self.idade} anos) criada.")

    # Mtodo de instncia
    def exibir_info(self) -> None:
        print(f"Nome: {self.nome_completo}, Idade: {self.idade}")

    # Mtodo de classe
    @classmethod
    def from_ano_nascimento(cls:'Pessoa', nome_completo:str, ano_nascimento:int) -> 'Pessoa':
        """
        Mtodo de classe para criar uma instncia de Pessoa a partir do ano de nascimento.
        'cls' refere-se  prpria classe Pessoa.
        """
        ano_atual = date.today().year
        idade = ano_atual - ano_nascimento
        print(f"  (Mtodo de classe: Calculando idade para {nome_completo}...)")
        return cls(nome_completo, idade)

    @classmethod
    def from_json(cls:'Pessoa', json_data:str) -> 'Pessoa':
        """
        Mtodo de classe para criar uma instncia de Pessoa a partir de dados JSON.
        'cls' refere-se  prpria classe Pessoa.
        """
        data:dict = json.loads(json_data)
        # return cls(data['nome_completo_da_pessoa'], data['idade'])
        return cls(**data)
````

````python
print("--- Criando Pessoas com __init__ padro ---")
pessoa1 = Pessoa(nome_completo="Maria Oliveira", idade=30)
pessoa1.exibir_info()

print("\n--- Criando Pessoas com mtodo de classe (from_data_nascimento) ---")
pessoa2 = Pessoa.from_ano_nascimento(nome_completo="Joo Costa", ano_nascimento=1990)
pessoa2.exibir_info()

print("\n--- Criando Pessoas com mtodo de classe (from json) ---")
pessoa3 = Pessoa.from_json(json_data='{"nome_completo" : "Jos da Silva", "idade" : 44}')
pessoa3.exibir_info()
````

## Mtodos estticos (`@staticmethod`)

Os **mtodos estticos** em Python so um tipo de mtodo que pertence  classe, mas no operam sobre a instncia da classe nem sobre a prpria classe. Eles so, em essncia, funes regulares que esto "encaixadas" dentro de uma classe, mas no recebem automaticamente a instncia (`self`) nem a classe (`cls`) como primeiro argumento.

Voc os define usando o decorador `@staticmethod` logo acima da sua definio de mtodo.

````python
class Produto:
    def __init__(self, nome:str, preco:float, ean:str) -> None:
        self.nome = nome
        self.preco = preco
        self.ean = Produto.calcular_digito_verificador(ean)

    def exibir_detalhes(self) -> None:
        print(f"Detalhes: {self.nome} - R${self.preco:.2f} Cdigo de barras '{self.ean}' criado.")

    @staticmethod
    def calcular_digito_verificador(ean:str) -> str:
        """
        Mtodo esttico para calcular o dgito verificador de um EAN-13.
        O EAN-13  composto por 12 dgitos + 1 dgito verificador.
        """
        if len(ean) != 12 or not ean.isdigit():
            raise ValueError("EAN deve ter 12 dgitos numricos.")

        soma_par = sum(int(ean[i]) for i in range(0, 12, 2))
        soma_impar = sum(int(ean[i]) for i in range(1, 12, 2))
        total = (soma_par * 3 + soma_impar) % 10
        digito_verificador = (10 - total) % 10
        return f"{ean}{digito_verificador}"
````

````python
produto1 = Produto(nome="Camiseta", preco=49.90, ean="123456189012")
produto1.exibir_detalhes()

ean_novo = Produto.calcular_digito_verificador(ean="987654327098")
print(f"EAN com dgito: {ean_novo}")
````

## Mtodos especiais: `__eq__`, `__lt__`, `__hash__`

Em Python, os mtodos especiais (frequentemente chamados de "**dunder methods**" por causa dos dois underscores __ que os envolvem, como `__eq__` para "**equal**") permitem que voc personalize o comportamento de suas classes quando elas interagem com operadores e funes built-in. Eles so a base para como os objetos se comportam em contextos como **comparao**, **ordenao**, **hashing** e at mesmo representao como string.

Vamos focar em trs mtodos cruciais para comparao e hashing: `__eq__`, `__lt__`, e `__hash__`.

1. `__eq__(self, other)`: Igualdade (==), define o comportamento do operador de igualdade (==) para objetos da sua classe. Por padro, se voc no definir `__eq__`, o Python compara objetos com base em sua identidade (se eles so o mesmo objeto na memria), o que raramente  o que voc quer para objetos com o mesmo valor.

````python
class Ponto:
    def __init__(self, x:float, y:float) -> None:
        self.x = x
        self.y = y

    def __str__(self) -> str:
        return f"Ponto({self.x}, {self.y})"

    def __eq__(self, other:'Ponto') -> bool:
        if not isinstance(other, Ponto):
            return NotImplemented
        return self.x == other.x and self.y == other.y
````

````python
p1 = Ponto(x=1, y=2)
p2 = Ponto(x=1, y=2)
p3 = Ponto(x=3, y=4)

print(f"{p1} == {p2}: {p1 == p2}")
print(f"{p1} == {p3}: {p1 == p3}")
print(f"{p2} == {p3}: {p2 == p3}")
print(f"{p1} == {p1}: {p1 == p1}")
````

2. `__lt__(self, other)`: Menor Que (<), define o comportamento do operador "menor que" (<). Se voc definir `__lt__`, pode ser possvel que outros operadores de comparao (como >, <=, >=) funcionem automaticamente (via functools.total_ordering ou se o Python puder inferir a relao).

**@functools.total_ordering**  um decorador que, se voc definir `__eq__` e **pelo menos um** dos mtodos de ordenao rica (`__lt__`, `__le__`, `__gt__`, `__ge__`), ele preenche automaticamente os outros mtodos de ordenao.

````python
from functools import total_ordering

@total_ordering
class Tarefa:
    def __init__(self, nome:str, prioridade:int) -> None:
        self.nome = nome
        self.prioridade = prioridade

    def __str__(self) -> str:
        return f"Tarefa '{self.nome}' (Prioridade: {self.prioridade})"

    def __eq__(self, other:'Tarefa') -> bool:
        if not isinstance(other, Tarefa):
            return NotImplemented
        return self.nome == other.nome and self.prioridade == other.prioridade

    def __lt__(self, other:'Tarefa') -> bool:
        if not isinstance(other, Tarefa):
            return NotImplemented
        return self.prioridade < other.prioridade
````

````python
t1 = Tarefa(nome="Fazer compras", prioridade=3)
t2 = Tarefa(nome="Responder e-mails", prioridade=1)
t3 = Tarefa(nome="Organizar documentos", prioridade=3)
t4 = Tarefa(nome="Ligar para cliente", prioridade=2)

print(f"{t2} < {t1}: {t2 < t1}")
print(f"{t1} < {t3}: {t1 < t3}")
print(f"{t3} > {t4}: {t3 > t4}")
print(f"{t1} <= {t3}: {t1 <= t3}")
print(f"{t1} > {t2}: {t1 > t2}")
````

3. `__hash__(self)`: Hashing (hash(), Sets, Dicionrios). Retorna um valor hash (um inteiro) para o objeto. Este valor  usado por estruturas de dados baseadas em hash, como sets (set) e dicionrios (dict), para armazenar e recuperar objetos eficientemente.

Para que um objeto seja hashable e funcione corretamente em sets e dicionrios, ele DEVE ser imutvel (ou pelo menos, seus atributos que contribuem para o hash devem ser imutveis e no devem mudar aps o objeto ser inserido em uma estrutura hashable).

Esse mtodo deve retornar um inteiro. Se dois objetos so considerados iguais (a == b  True), ento seus valores hash **DEVEM** ser iguais (hash(a) == hash(b) tambm deve ser True). Se voc no definir `__hash__` (e definir `__eq__`), seus objetos se tornaro no hashable por padro (a menos que a superclasse defina um `__hash__`).

````python
class Coordenada:
    def __init__(self, x:float, y:float) -> None:
        self.x = x
        self.y = y

    def __str__(self) -> str:
        return f"Coordenada({self.x}, {self.y})"

    def __repr__(self) -> str:
        return f"Coordenada(x={self.x}, y={self.y})"

    def __eq__(self, other:'Coordenada') -> bool:
        if not isinstance(other, Coordenada):
            return NotImplemented
        return self.x == other.x and self.y == other.y

    def __hash__(self) -> int:
        # Gerar um hash baseado nos atributos x e y
        # A funo hash() pode ser aplicada a tuplas, que so hashable.
        # Uma boa prtica  combinar os hashes dos atributos imutveis.
        return hash((self.x, self.y))
````

## Decorators de funes

Um decorator (`@`) em Python  uma funo que recebe uma funo como argumento e retorna uma nova funo (geralmente uma verso modificada da funo original). Eles so uma forma elegante e reutilizvel de adicionar funcionalidades a funes ou mtodos existentes sem modificar diretamente o cdigo dessas funes.

Pense em um decorador como um **"embrulho"** ou um **"invlucro"** em torno de uma funo. Ele permite que voc execute cdigo antes e/ou depois da execuo da funo original.

````python
from typing import Callable, Any

def meu_decorador_simples(funcao:Callable):
    def wrapper() -> Any:
        print("Antes da funo ser chamada.")
        funcao() # Chama a funo original
        print("Depois da funo ser chamada.")
    return wrapper
````

````python
@meu_decorador_simples
def ola_mundo() -> None:
    print("Ol, mundo!")

ola_mundo()
````

## Decorators de mtodos e classes

Anteriormente, falamos sobre decorators de funes, que operam sobre funes avulsas. Agora, vamos estender esse conceito para decorators de mtodos (que so um tipo de funo) e decorators de classes.

1. **Decorators de Mtodos**: Um decorator de mtodo , essencialmente, um decorator de funo que  aplicado a um mtodo dentro de uma classe. Ele funciona da mesma forma: recebe o mtodo original e retorna um novo mtodo (o "embrulho").

````python
def log_metodo(func:Callable):
    """
    Decorador para registrar a entrada e sada de um mtodo.
    """
    def wrapper(self, *args, **kwargs) -> Any | Exception:

        # O 'self'  crucial aqui para mtodos de instncia
        nome_classe:str = self.__class__.__name__
        nome_metodo:str = func.__name__
        print(f"@ Entrando em {nome_classe}.{nome_metodo} com args: {args}, kwargs: {kwargs}")

        try:
            # Chama o mtodo original
            resultado:Any = func(self, *args, **kwargs)
            print(f"@ Saindo de {nome_classe}.{nome_metodo}. Resultado: {resultado}")
            return resultado
        except Exception as e:
            print(f"@ Erro em {nome_classe}.{nome_metodo}: {e}")
            raise e

    return wrapper
````

````python
class ContaBancaria:
    def __init__(self, numero_conta:str, saldo_inicial:float=0.0):
        self.numero_conta = numero_conta
        self.saldo = saldo_inicial
        print(f"Conta {self.numero_conta} criada com saldo R${self.saldo:.2f}")

    @log_metodo
    def depositar(self, valor:float) -> float | ValueError:
        if valor <= 0:
            raise ValueError("Valor de depsito deve ser positivo.")
        self.saldo += valor
        print(f"Depsito de R${valor:.2f} na conta {self.numero_conta}.")
        return self.saldo

    @log_metodo
    def consultar_saldo(self) -> float:
        print(f"Consultando saldo da conta {self.numero_conta}.")
        return self.saldo
````

````python
print("--- Demonstrao de Decorators de Mtodos ---")
minha_conta = ContaBancaria(numero_conta="12345", saldo_inicial=1000.00)

minha_conta.depositar(valor=500.00)
saldo = minha_conta.consultar_saldo()
print(f"Saldo final da conta {minha_conta.numero_conta}: R${saldo:.2f}")

try:
    minha_conta.depositar(valor=-100.00)
except ValueError as e:
    print(f"Erro ao depositar: {e}")
````

2. **Decorators de Classes**: Um decorator de classe  uma funo que recebe uma classe como argumento e retorna uma nova classe (geralmente uma verso modificada da classe original). Eles so usados para modificar ou estender o comportamento de uma classe inteira.

````python
def adicionar_info_versao(versao:str="1.0"):
    """
    Funo de fbrica de decoradores que adiciona um atributo de classe 'VERSAO'
    e um mtodo de classe 'get_versao'  classe decorada.
    """
    def wrapper(cls):
        # 'cls' aqui  a prpria classe que est sendo decorada
        # Adiciona um atributo de classe
        cls.VERSAO = versao

        # Adiciona um mtodo de classe  classe decorada
        @classmethod
        def get_versao(cls):
            return cls.VERSAO

        # Atribui o novo mtodo  classe
        cls.get_versao = get_versao

        print(f"Decorador: Adicionado VERSAO '{versao}' e mtodo 'get_versao'  classe {cls.__name__}")

        # Retorna a classe modificada
        return cls

    return wrapper
````

````python
@adicionar_info_versao(versao="2.0-beta")
class MeuSistema:
    def __init__(self, nome:str) -> None:
        self.nome = nome
        print(f"Sistema '{self.nome}' inicializado.")

    def iniciar(self) -> None:
        print(f"Iniciando {self.nome} (Verso: {self.get_versao()})...")
````

````python
sistema = MeuSistema(nome="Sistema de Gerenciamento")
sistema.iniciar()

print(f"Verso do sistema via atributo: {MeuSistema.VERSAO}")
print(f"Verso do sistema via mtodo: {MeuSistema.get_versao()}")
````

## functools.wraps

Quando voc cria um decorator em Python, ele essencialmente **"embrulha"** a funo original, substituindo-a por uma nova funo (o wrapper). Embora isso seja poderoso para adicionar funcionalidades, tem uma desvantagem: a funo decorada **perde sua identidade original**. Informaes importantes como o nome da funo (`__name__`), seu mdulo (`__module__`), seu docstring (`__doc__`), e seus argumentos (`__annotations__` ou `__signature__`) so substitudas pelas informaes do wrapper e no da funo.

 aqui que entra o `functools.wraps`. Ele  um decorador que voc usa dentro do seu prprio decorador para **preservar os metadados** da funo original. Ele copia os atributos relevantes da funo embrulhada para a funo wrapper, tornando o wrapper "parecido" com a funo original.

````python
# Exemplo SEM o uso do functools.wraps
def meu_decorador_sem_wraps(func:Callable):
    def wrapper(*args, **kwargs) -> Any:
        """DocString do Wrapper"""
        return func(*args, **kwargs)
    return wrapper


@meu_decorador_sem_wraps
def soma(a:int, b:int) -> int:
    """Soma dois nmeros."""
    return a + b

print("Sada:", soma.__name__) # Esperado o nome da funo soma
print("Sada:", soma.__doc__)  # Esperado o docstring da funo "Soma dois nmeros."
````

````python
from functools import wraps

def meu_decorador_com_wraps(func:Callable):
    @wraps(func)
    def wrapper(*args, **kwargs) -> Any:
        """DocString do Wrapper"""
        return func(*args, **kwargs)
    return wrapper

@meu_decorador_com_wraps
def soma(a:int, b:int) -> int:
    """Soma dois nmeros."""
    return a + b

print("Sada:", soma.__name__) # Esperado o nome da funo soma
print("Sada:", soma.__doc__)  # Esperado o docstring da funo "Soma dois nmeros."
````

## Decorators com argumentos

Anteriormente, vimos que um decorator  uma funo que recebe outra funo e a **"embrulha"** para adicionar funcionalidade. Mas e se o prprio decorator precisar de configuraes ou parmetros?  a que entram os decorators com argumentos.

Para que um decorator aceite argumentos, ele precisa ser implementado como uma **funo de fbrica** (factory function). Isso significa que o decorator em si  uma funo que retorna o verdadeiro decorator.

````python
def validar_retorno(tipo_esperado:Any): # Factory Function
    """
    Funo de fbrica de decoradores.
    Retorna um decorador que valida o tipo do valor de retorno de uma funo.
    """
    # Esta  a funo que DECORA 'func'
    def decorador_real(func:Callable):

        @wraps(func)
        def wrapper(*args, **kwargs) -> Any:
            # Executa a funo original
            resultado = func(*args, **kwargs)

            # Validao do tipo de retorno
            if not isinstance(resultado, tipo_esperado):
                raise TypeError(
                    f"O retorno da funo '{func.__name__}'  do tipo '{type(resultado).__name__}', "
                    f"mas o esperado era '{tipo_esperado.__name__}'."
                )
            print(f"Sucesso: Retorno de '{func.__name__}'  do tipo '{tipo_esperado.__name__}'.")
            return resultado

        return wrapper

    return decorador_real
````

````python
@validar_retorno(int) # O decorator espera que 'somar' retorne um inteiro
def somar(a:int, b:int) -> int:
    """Soma dois nmeros inteiros."""
    return a + b

soma_resultado = somar(a=5, b=3)
print(f"Resultado da soma: {soma_resultado}\n\n")

try:
    soma_resultado = somar(a=5.0, b=3)
    print(f"Resultado da soma: {soma_resultado}")
except TypeError as e:
    print(f"Erro na soma: {e}")
````

## Classes como decorators

At agora, vimos que decorators so funes que embrulham outras funes. No entanto, em Python, voc tambm pode implementar um decorator como uma classe. Usar uma classe como decorator  particularmente til quando voc precisa que o decorator mantenha algum estado ou precise de mais complexidade na lgica de "embrulho".

Para que uma classe funcione como um decorator, ela precisa ter:

1. Um mtodo `__init__(self, func)`: Este mtodo ser chamado quando o decorator for aplicado  funo. Ele receber a funo a ser decorada como seu primeiro argumento. A classe decorator geralmente armazena essa funo.

2. Um mtodo `__call__(self, *args, **kwargs)`: Este mtodo especial transforma a instncia da classe em um **callable** (algo que pode ser chamado como uma funo). Quando a funo decorada  chamada, na verdade  este mtodo `__call__` da instncia do decorator que  invocado. Dentro de `__call__`, voc executa a lgica do decorator e chama a funo original.

````python
class ContadorChamadas:
    """
    Um decorator de classe que conta quantas vezes uma funo foi chamada.
    """
    def __init__(self, func:Callable) -> None:
        self.func = func

        # Estado: contador de chamadas
        self.num_chamadas:int = 0

        # Preserva metadados
        # @wraps(func) -> wrapper(contexto)
        wraps(self.func)(self)


    def __call__(self, *args, **kwargs) -> Any:
        # Incrementa o contador
        self.num_chamadas += 1

        print(f"--> Funo '{self.func.__name__}' chamada {self.num_chamadas} vez(es).")

        # Chama a funo original
        resultado = self.func(*args, **kwargs)
        return resultado
````

````python
@ContadorChamadas
def executar_tarefa(tarefa_id:int) -> str:
    """Simula a execuo de uma tarefa."""
    print(f"Executando tarefa ID: {tarefa_id}")
    return f"Tarefa {tarefa_id} concluda."


resultado_1 = executar_tarefa(tarefa_id=1)
resultado_2 = executar_tarefa(tarefa_id=2)
resultado_3 = executar_tarefa(tarefa_id=1)
````

## Introduo a metaclasses

As **metaclasses** em Python so um dos conceitos mais avanados e poderosos (e por vezes complexos) da linguagem. Em sua essncia, uma metaclasse  a classe de uma classe.

Pense assim:
* Um objeto  uma instncia de uma classe.
* Uma classe  uma instncia de uma metaclasse.

A metaclasse padro em Python, da qual todas as classes (que voc cria ou as built-in) so instncias,  `type`.

Metaclasses so usadas para controlar a criao de classes. Se voc precisa modificar a forma como as classes so definidas, ou injetar comportamento nas classes no momento em que elas so criadas (no quando suas instncias so criadas), as metaclasses so a ferramenta.

Para definir uma metaclasse, voc cria uma classe que herda de `type` (ou de uma subclasse de type). O mtodo mais comum e importante que voc sobrescreve em uma metaclasse  `__new__`.

`__new__(mcs, name, bases, attrs)`: Este mtodo  responsvel por criar e retornar o novo objeto classe.

* **mcs**: A prpria metaclasse (anlogo a cls em um classmethod).
* **name**: O nome da classe que est sendo criada (ex: "MinhaClasse").
* **bases**: Uma tupla das classes base (superclasses) da nova classe.
* **attrs**: Um dicionrio contendo os atributos e mtodos da nova classe (ex: {'atributo': 10, 'metodo': <function ...>}).



````python
class MinhaMetaclasse(type):
    def __new__(mcs:'MinhaMetaclasse', name:str, bases:tuple, attrs:dict):
        # 1. Adiciona um atributo padro ao dicionrio de atributos
        if 'created_by' not in attrs:
            attrs['created_by'] = "Metaclasse Padro"

        # 2. Cria a classe usando o __new__ da metaclasse base (type)
        nova_classe = super().__new__(mcs, name, bases, attrs)

        # 3. Opcional: Adicionar lgica ps-criao da classe
        print(f"Metaclasse: Classe '{name}' criada. Atributo 'created_by' adicionado.")

        return nova_classe
````

````python
class Produto(metaclass=MinhaMetaclasse):
    def __init__(self, nome):
        self.nome = nome

    def exibir_info(self):
        print(f"Produto: {self.nome}, Criado por: {self.created_by}")

prod = Produto(nome="Televisor")
prod.exibir_info()

print(f"Tipo de Produto: {type(Produto)}")
````

Metaclasses so uma ferramenta poderosa, mas devem ser usadas com cautela e moderao. Elas adicionam uma camada de complexidade e podem dificultar a leitura do cdigo se no forem usadas de forma transparente.

**Considere usar metaclasses quando:**

* Voc precisa modificar o comportamento das classes em si, no apenas das instncias.

* Voc precisa impor polticas ou validaes rigorosas sobre como as classes so definidas.

* Outras solues (como decorators de classe ou funes simples) no so suficientes para o problema.

Na maioria dos casos do dia-a-dia, voc no precisar criar suas prprias metaclasses. O `ABC` para classes abstratas e decorators de classe para manipulaes mais simples geralmente so suficientes. No entanto, entender metaclasses aprofunda sua compreenso de como o Python constri suas classes e objetos.

## Singleton

![singleton](https://refactoring.guru/images/patterns/cards/singleton-mini.png)\
Fonte: https://refactoring.guru/

\

O padro de **design Singleton**  um padro de design criacional que garante que uma classe tenha apenas uma instncia e fornece um ponto de acesso global a essa instncia. Em outras palavras, no importa quantas vezes voc tente criar um objeto dessa classe, voc sempre receber a mesma e **nica instncia**.

````python
class ConfigManager:

    # Atributo de classe para armazenar a nica instncia
    _instance = None
    # Dicionrio para armazenar as configuraes
    _configs = {}

    def __new__(cls):
        # Se a instncia ainda no existe, cria uma nova
        if cls._instance is None:
            print("ConfigManager: Criando a PRIMEIRA e NICA instncia.")
            cls._instance = super().__new__(cls)
            cls._configs = {
                "DATABASE_URL": "sqlite:///app.db",
                "API_KEY": "default_api_key_123",
                "LOG_LEVEL": "INFO"
            }
        else:
            print("ConfigManager: Retornando a instncia j existente.")
        return cls._instance

    def get_config(self, key):
        return self._configs.get(key)

    def set_config(self, key, value):
        self._configs[key] = value

    def show_all_configs(self):
        print("\n--- Configuraes Atuais ---")
        for key, value in self._configs.items():
            print(f"  {key}: {value}")
        print("--------------------------")
````

````python
config1 = ConfigManager() # A instncia  criada aqui

print("config1.show_all_configs()")
config1.show_all_configs()

config2 = ConfigManager() # A mesma instncia  retornada aqui

print(f"config1  o mesmo objeto que config2? {config1 is config2}")

print("config1.set_config")
config1.set_config("LOG_LEVEL", "DEBUG")

print("config2.show_all_configs()")
config2.show_all_configs()
print("config2.set_config")
config2.set_config("SECRET_KEY", "minha_chave_secreta_abc")

print("config1.show_all_configs()")
config1.show_all_configs()
````

## Factory Method

![Factory](https://refactoring.guru/images/patterns/cards/factory-method-mini.png) \
Fonte: https://refactoring.guru/

\

O padro de **design Factory Method** (Mtodo de Fbrica)  um padro de design criacional que define uma interface para criar um objeto, mas permite que as subclasses decidam qual classe instanciar. Em outras palavras, ele delega a responsabilidade de instanciar objetos para as subclasses.

O Factory Method visa resolver o problema de criar objetos sem especificar a classe exata do objeto que ser criado. Ele oferece uma forma flexvel de criar diferentes tipos de objetos relacionados sem expor a lgica de criao complexa ao cliente (o cdigo que usa a fbrica).

````python
from abc import ABC, abstractmethod
from datetime import datetime

class Logger(ABC):
    """
    Interface abstrata para todos os tipos de loggers.
    """
    @abstractmethod
    def log(self, mensagem:str) -> None:
        pass

    def _formatar_mensagem(self, mensagem:str) -> str:
        """Mtodo comum para formatar mensagens de log."""
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        return f"[{timestamp}] {mensagem}"
````

````python
class ConsoleLogger(Logger):
    """
    Implementao de logger que escreve mensagens no console.
    """
    def log(self, mensagem:str) -> None:
        mensagem_formatada = self._formatar_mensagem(mensagem)
        print(f"CONSOLE: {mensagem_formatada}")
````

````python
class FileLogger(Logger):
    """
    Implementao de logger que escreve mensagens em um arquivo.
    """
    def __init__(self, nome_arquivo:str="app.log") -> None:
        self.nome_arquivo = nome_arquivo
        print(f"FileLogger: Mensagens sero gravadas em '{self.nome_arquivo}'.")

    def log(self, mensagem:str) -> None:
        mensagem_formatada = self._formatar_mensagem(mensagem)
        with open(self.nome_arquivo, 'a') as f:
            f.write(f"FILE: {mensagem_formatada}\n")
        print(f"FileLogger: Log gravado em arquivo: {mensagem_formatada}")
````

````python
from typing import Type

class LoggerFactory(ABC):
    """
    Classe abstrata que declara o Factory Method.
    Pode conter lgica comum que usa o logger criado.
    """

    @abstractmethod
    def criar_logger(self) -> Type[Logger]:
        """
        Mtodo de fbrica abstrato.
        Subclasses devem implement-lo para retornar um tipo especfico de Logger.
        """
        pass

    def registrar_evento(self, evento:str) -> None:
        """
        Mtodo que usa o logger criado pelo Factory Method.
        No precisa saber qual tipo de logger est usando.
        """
        # Chama o Factory Method para obter o logger
        logger = self.criar_logger()

        logger.log(f"Evento registrado: {evento}")
        print(f"LoggerFactory: Evento '{evento}' processado.")
````

````python
class ConsoleLoggerFactory(LoggerFactory):
    """
    Fbrica que cria instncias de ConsoleLogger.
    """
    def criar_logger(self) -> Type[Logger]:
        return ConsoleLogger()
````

````python
class FileLoggerFactory(LoggerFactory):
    """
    Fbrica que cria instncias de FileLogger.
    """
    def __init__(self, nome_arquivo:str="default.log") -> None:
        self.nome_arquivo = nome_arquivo

    def criar_logger(self) -> Type[Logger]:
        return FileLogger(self.nome_arquivo)
````

````python
console_factory = ConsoleLoggerFactory()
console_factory.registrar_evento(evento="Usurio fez login")
console_factory.registrar_evento(evento="Item adicionado ao carrinho")
````

````python
file_factory = FileLoggerFactory(nome_arquivo="meus_logs.log")
file_factory.registrar_evento(evento="Erro na conexo com o banco de dados")
file_factory.registrar_evento(evento="Backup de dados concludo")
````

**Vantagens do Factory Method**:

* **Separao de Preocupaes**: O cdigo cliente  desacoplado da lgica de criao dos produtos concretos. Ele opera apenas com a interface do Criador e do Produto.

* **Extensibilidade**: Adicionar um novo tipo de logger (e.g., DatabaseLogger)  fcil. Basta criar uma nova classe DatabaseLogger (Produto Concreto) e uma nova DatabaseLoggerFactory (Criador Concreto). O cdigo cliente existente no precisar ser modificado.

* **Flexibilidade**: Voc pode alternar entre diferentes tipos de produtos concretos sem alterar o cdigo que os utiliza.

O **Factory Method**  um padro poderoso para lidar com a criao de objetos em sistemas onde a flexibilidade e a extensibilidade so importantes.

## Strategy

![Strategy](https://refactoring.guru/images/patterns/cards/strategy-mini.png)\
Fonte: https://refactoring.guru/

\

O padro de **design Strategy** (Estratgia)  um padro de design comportamental que permite definir uma famlia de algoritmos, encapsular cada um deles e torn-los intercambiveis. Ele permite que o algoritmo varie independentemente dos clientes que o utilizam.

Pense nisso como ter vrias maneiras de fazer a mesma coisa (por exemplo, diferentes formas de pagar, diferentes formas de classificar dados, diferentes formas de compactar um arquivo), e voc pode trocar qual mtodo usar em tempo de execuo.

````python
class CalculadorFrete(ABC):
    """
    Interface para algoritmos de clculo de frete.
    """
    @abstractmethod
    def calcular(self, peso_kg:float, distancia_km:float) -> float:
        """
        Calcula o valor do frete.
        """
        pass
````

````python
class FreteCorreiosNormal(CalculadorFrete):
    """
    Estratgia para clculo de frete via Correios Normal.
    """
    def calcular(self, peso_kg:float, distancia_km:float) -> float:
        taxa_base = 10.0
        custo_por_kg = 2.50
        custo_por_km = 0.10
        return taxa_base + (peso_kg * custo_por_kg) + (distancia_km * custo_por_km)
````

````python
class FreteSedex(CalculadorFrete):
    """
    Estratgia para clculo de frete via Sedex (mais caro, mas rpido).
    """
    def calcular(self, peso_kg:float, distancia_km:float) -> float:
        taxa_base = 25.0
        custo_por_kg = 5.00
        custo_por_km = 0.25
        return taxa_base + (peso_kg * custo_por_kg) + (distancia_km * custo_por_km)
````

````python
class FreteTransportadoraExpressa(CalculadorFrete):
    """
    Estratgia para clculo de frete via Transportadora Expressa
    (fixo para pequenas distncias).
    """
    def calcular(self, peso_kg:float, distancia_km:float) -> float:
        if distancia_km < 50:
            return 30.0 # Valor fixo para curtas distncias
        else:
            taxa_base = 40.0
            custo_por_kg = 4.0
            custo_por_km = 0.15
            return taxa_base + (peso_kg * custo_por_kg) + (distancia_km * custo_por_km)
````

````python
class Pedido:
    """
    A classe Pedido usa uma estratgia de clculo de frete.
    """
    def __init__(self, peso_kg:float, distancia_km:float, calculador_frete: Type[CalculadorFrete]) -> None:
        self.peso_kg = peso_kg
        self.distancia_km = distancia_km
        self._calculador_frete = calculador_frete

    def set_calculador_frete(self, novo_calculador: Type[CalculadorFrete]) -> None:
        """Permite mudar a estratgia de frete em tempo de execuo."""
        print(f"Mudando estratgia de frete para: {novo_calculador.__class__.__name__}")
        self._calculador_frete = novo_calculador

    def calcular_frete_total(self) -> float:
        """
        Delega o clculo do frete para a estratgia atual.
        """
        print(f"Calculando frete para Peso: {self.peso_kg}kg, Distncia: {self.distancia_km}km")
        return self._calculador_frete.calcular(self.peso_kg, self.distancia_km)
````

````python
print("--- Cenrio 1: Pedido com Frete Correios Normal ---")
# Cliente escolhe a estratgia na criao do pedido
pedido1 = Pedido(
    peso_kg=5,
    distancia_km=100,
    calculador_frete=FreteCorreiosNormal()
)
valor_frete1 = pedido1.calcular_frete_total()
print(f"Valor do frete (Correios Normal): R${valor_frete1:.2f}\n")


print("--- Cenrio 2: Pedido com Frete Sedex ---")
pedido2 = Pedido(
    peso_kg=2,
    distancia_km=50,
    calculador_frete=FreteSedex()
)
valor_frete2 = pedido2.calcular_frete_total()
print(f"Valor do frete (Sedex): R${valor_frete2:.2f}\n")

print("--- Cenrio 3: Mudando a Estratgia em Tempo de Execuo ---")
pedido3 = Pedido(
    peso_kg=8,
    distancia_km=30,
    calculador_frete=FreteCorreiosNormal() # Comea com Correios Normal
)
valor_frete3_inicial = pedido3.calcular_frete_total()
print(f"Valor do frete inicial (Correios Normal): R${valor_frete3_inicial:.2f}")

# Mudando a estratgia para Transportadora Expressa
pedido3.set_calculador_frete(novo_calculador=FreteTransportadoraExpressa())
valor_frete3_novo = pedido3.calcular_frete_total()
print(f"Valor do frete novo (Transportadora Expressa): R${valor_frete3_novo:.2f}")
````

## Observer

![Observer](https://refactoring.guru/images/patterns/cards/observer-mini.png)\
Fonte: https://refactoring.guru/

\

O padro de **design Observer** (Observador)  um padro de design comportamental que define uma dependncia um-para-muitos entre objetos, de modo que, quando um objeto muda de estado, todos os seus dependentes so notificados e atualizados automaticamente.

Imagine uma assinatura de newsletter: voc se inscreve (se torna um observador) e, sempre que h uma nova edio (mudana de estado no editor), voc a recebe ( notificado).

O Observer visa definir uma forma de comunicao entre objetos de forma que o objeto que tem o estado (o Sujeito ou Observvel) no precise saber os detalhes de quem est interessado nesse estado (os Observadores). Isso promove um acoplamento fraco entre os objetos, tornando o sistema mais flexvel e fcil de manter e estender.

````python
# Observador (Quem ser avisado)
class Assinante(ABC):
    """
    Define a interface para objetos que so notificados sobre mudanas.
    """
    @abstractmethod
    def atualizar(self, noticia):
        """
        Recebe a atualizao do Sujeito.
        """
        pass
````

````python
# Observvel (Quem estamos aguardando)
class AgenciaDeNoticias:
    """
    Mantm uma lista de assinantes e os notifica sobre novas notcias.
    """
    def __init__(self):
        # Lista de observadores
        self._assinantes = []
        self._ultima_noticia = None

    def adicionar_assinante(self, assinante: Type[Assinante]):
        """Anexa um observador."""
        if assinante not in self._assinantes:
            self._assinantes.append(assinante)
            print(f"Agencia: Assinante '{assinante.__class__.__name__}' adicionado.")

    def remover_assinante(self, assinante: Type[Assinante]):
        """Desanexa um observador."""
        if assinante in self._assinantes:
            self._assinantes.remove(assinante)
            print(f"Agencia: Assinante '{assinante.__class__.__name__}' removido.")

    def __notificar_assinantes(self):
        """Notifica todos os observadores sobre a ltima notcia."""
        print(f"\nAgencia: Notificando {len(self._assinantes)} assinante(s) sobre a nova notcia...")
        for assinante in self._assinantes:
            # Chama o mtodo de atualizao do observador
            assinante.atualizar(self._ultima_noticia)

    def publicar_noticia(self, titulo, conteudo):
        """Muda o estado do Sujeito e notifica os observadores."""
        self._ultima_noticia = {"titulo": titulo, "conteudo": conteudo}
        print(f"\nAgencia: Notcia Publicada! '{titulo}'")
        self.__notificar_assinantes()
````

````python
# Observador Concreto
class LeitorWeb(Assinante):
    """
    Assinante que simula a atualizao de um site ou aplicativo.
    """
    def __init__(self, nome):
        self.nome = nome
        self.noticias_recebidas = []
        print(f"Leitor Web '{self.nome}' criado.")

    def atualizar(self, noticia):
        self.noticias_recebidas.append(noticia)
        print(f"  Leitor Web {self.nome}: Notcia '{noticia['titulo']}' recebida. Atualizando feed!")
        print(f"    Contedo: {noticia['conteudo'][:50]}...")
````

````python
# Observador Concreto
class EmailAssinante(Assinante):
    """
    Assinante que simula o recebimento de e-mails de notcias.
    """
    def __init__(self, email):
        self.email = email
        self.emails_recebidos = 0
        print(f"Email Assinante '{self.email}' criado.")

    def atualizar(self, noticia):
        self.emails_recebidos += 1
        print(f"  Email Assinante {self.email}: Enviando e-mail com a notcia '{noticia['titulo']}'.")
        print(f"    (Total de e-mails enviados: {self.emails_recebidos})")
````

````python
agencia = AgenciaDeNoticias()

assinante_web1 = LeitorWeb("Alice")
assinante_web2 = LeitorWeb("Bob")
assinante_email1 = EmailAssinante("contato@exemplo.com")

# Adicionando assinantes  agncia
agencia.adicionar_assinante(assinante_web1)
agencia.adicionar_assinante(assinante_email1)
agencia.adicionar_assinante(assinante_web2)

# Nova notcia
agencia.publicar_noticia(titulo="Lanamento do Novo Smartphone X",
                         conteudo="A TechCo anunciou hoje seu revolucionrio Smartphone X, prometendo mudar o mercado...")

agencia.remover_assinante(assinante_web1)

# Outra notcia
agencia.publicar_noticia(titulo="Economia Global em Recuperao",
                         conteudo="Relatrio recente indica crescimento otimista do PIB em diversas regies...")
````

## Injeo de dependncia bsica

A **Injeo de Dependncia** (DI) no  um padro de design no mesmo sentido que Singleton ou Factory Method (que so padres criacionais). Em vez disso,  um princpio de design e uma tcnica que visa a inverso de controle (`IoC`), especificamente no que diz respeito s dependncias de um objeto.

Tradicionalmente, uma classe  responsvel por criar ou localizar todas as suas dependncias. Isso leva a um **acoplamento forte**.

````python
# Sem DI
class BancoDeDados:
    def conectar(self):
        print("Conectando ao banco de dados MySQL...")

class RepositorioUsuario:
    def __init__(self):
        # RepositorioUsuario CRIA sua prpria dependncia
        self.db = BancoDeDados()

    def buscar_usuario(self, id):
        self.db.conectar()
        print(f"Buscando usurio {id} no banco de dados.")
        # ... lgica de busca
````

**Problemas com o cdigo acima**:

`RepositorioUsuario` est rigidamente acoplado a `BancoDeDados`. Se voc quiser usar outro tipo de banco (PostgreSQL, MongoDB), ter que modificar RepositorioUsuario.

**Difcil de testar**: Para testar `RepositorioUsuario`, voc sempre precisa de um `BancoDeDados` real, o que torna os testes lentos e dependentes de um ambiente externo.

**A Soluo da Injeo de Dependncia**:

Em vez de uma classe criar suas dependncias, as dependncias so fornecidas (injetadas) a ela de fora. O controle de quem cria e gerencia essas dependncias  invertido, passando para um "injetor" ou continer.

**Princpios da DI**:
* **Inverso de Controle (IoC)**: O controle da criao e gerenciamento das dependncias  invertido. A classe no controla mais suas dependncias; elas so dadas a ela.

* **Acoplamento Fraco**: As classes se tornam menos dependentes de implementaes concretas, dependendo apenas de interfaces (classes abstratas) ou de tipos genricos.

* **Testabilidade**: Facilita a substituio de dependncias reais por "mocks" ou "stubs" em testes.

**Tipos Bsicos de Injeo de Dependncia**:

Existem trs formas principais de injetar dependncias:

1. **Injeo por Construtor** (Constructor Injection)
Como funciona: As dependncias so passadas como argumentos para o mtodo construtor (`__init__`) da classe.  a forma mais comum e recomendada, pois garante que o objeto seja criado em um estado vlido, com todas as suas dependncias essenciais. Isso garante que todas as dependncias necessrias estejam disponveis desde o momento da criao do objeto.

````python
class DBConnection(ABC):
    @abstractmethod
    def connect(self) -> None:
        pass

    @abstractmethod
    def disconnect(self) -> None:
        pass

    @abstractmethod
    def execute_query(self, query:str) -> dict:
        pass
````

````python
class MySQLConnection(DBConnection):
    def connect(self) -> None:
        print("MySQL: Conectado ao banco de dados.")

    def disconnect(self) -> None:
        print("MySQL: Desconectado do banco de dados.")

    def execute_query(self, query:str) -> dict:
        print(f"MySQL: Executando query: '{query}'")
        return {"data": f"Resultado de '{query}' do MySQL"}


class PostgreSQLConnection(DBConnection):
    def connect(self) -> None:
        print("PostgreSQL: Conectado ao banco de dados Pg.")

    def disconnect(self) -> None:
        print("PostgreSQL: Desconectado do banco de dados Pg.")

    def execute_query(self, query:str) -> dict:
        print(f"PostgreSQL: Executando query: '{query}'")
        return {"data": f"Resultado de '{query}' do PostgreSQL"}
````

````python
class UserRepository:

    # A dependncia (db_connection)  INJETADA via construtor
    def __init__(self, db_connection: Type[DBConnection]) -> None:
        self.db = db_connection
        print(f"UserRepository: Inicializado com {type(db_connection).__name__}.")

    def get_user_by_id(self, user_id:int):
        self.db.connect()
        result = self.db.execute_query(f"SELECT * FROM users WHERE id = {user_id}")
        self.db.disconnect()
        return result
````

````python
# Usando MySQL
mysql_conn = MySQLConnection()
user_repo_mysql = UserRepository(db_connection=mysql_conn)
user_data_mysql = user_repo_mysql.get_user_by_id(user_id=1)
print(f"Dados do usurio (MySQL): {user_data_mysql}")

print("\n")

# Usando PostgreSQL (sem mudar o cdigo de UserRepository)
pg_conn = PostgreSQLConnection()
user_repo_pg = UserRepository(db_connection=pg_conn)
user_data_pg = user_repo_pg.get_user_by_id(user_id=2)
print(f"Dados do usurio (PostgreSQL): {user_data_pg}")
````

2. **Injeo por Setter** (Setter Injection): A dependncia  fornecida atravs de um mtodo "setter" pblico (e.g., set_db_connection). Isso permite que as dependncias sejam configuradas aps a criao do objeto, til para dependncias opcionais ou que podem mudar em tempo de execuo. O objeto pode existir em um estado invlido se as dependncias essenciais no forem injetadas antes do uso.

````python
class Mailer(ABC):
    @abstractmethod
    def send_email(self, to:str, subject:str, body:str) -> None:
        pass

class SMTPServer(Mailer):
    def send_email(self, to:str, subject:str, body:str) -> None:
        print(f"SMTP: Enviando '{subject}' para {to}")

class MockMailer(Mailer):
    def send_email(self, to:str, subject:str, body:str) -> None:
        print(f"MockMailer: Simula envio de '{subject}' para {to}")
````

````python
class EmailService:
    def __init__(self):
        self.__mailer = None # Dependncia opcional ou configurvel posteriormente
        print("EmailService: Inicializado (mailer ainda no definido).")

    def set_mailer(self, mailer:Type[Mailer]) -> None:
        self.__mailer = mailer
        print(f"EmailService: Mailer definido para {type(mailer).__name__}.")

    def send_welcome_email(self, recipient_email:str) -> None:
        if self.__mailer:
            print(f"EmailService: Enviando e-mail de boas-vindas para {recipient_email} usando {type(self.__mailer).__name__}.")
            self.__mailer.send_email(recipient_email, "Bem-vindo!", "Obrigado por se registrar!")
        else:
            print("EmailService: Erro - Mailer no configurado.")
````

````python
print("\n--- Injeo por Setter ---")
email_service = EmailService()

smtp_mailer = SMTPServer()
email_service.set_mailer(mailer=smtp_mailer)
email_service.send_welcome_email("usuario@dominio.com")

print("\n--- Mudando o Mailer ---")
mock_mailer = MockMailer()
email_service.set_mailer(mailer=mock_mailer)
email_service.send_welcome_email("outro@dominio.com")

print("\n--- ERRO sem o Mailer ---")
email_service_2 = EmailService()
email_service_2.send_welcome_email("sem_mailer@dominio.com")
````

3. **Injeo por Atributo** (Property Injection): A dependncia  atribuda diretamente a um atributo pblico do objeto (e.g., minha_classe.dependencia = obj).  a forma mais simples, mas tambm a menos controlada. Cuidado, pois o objeto pode ser usado em um estado invlido se as dependncias essenciais no forem atribudas. Menos explcito do que o construtor ou setter.

````python
class ServerSender(ABC):
    @abstractmethod
    def send(self, recipient:str, msg:str) -> None:
        """Envia uma mensagem para o destinatrio."""
        pass

class SMSGateway(ServerSender):
    def send(self, recipient:str, msg:str) -> None:
        print(f"SMSGateway: Enviando SMS para {recipient}: {msg}")

class PushNotificationService(ServerSender):
    def send(self, recipient:str, msg:str) -> None:
        print(f"PushService: Enviando Push para {recipient}: {msg}")
````

````python
class Notifier:
    def __init__(self):
        self.sender = None # Dependncia que ser injetada por atributo PBLICO
        print("Notifier: Inicializado.")

    def notify_user(self, user_id:str, message:str) -> None:
        if self.sender:
            print(f"Notifier: Enviando notificao para {user_id} via {type(self.sender).__name__}: {message}")
            self.sender.send(user_id, message)
        else:
            print("Notifier: Erro - Sender no definido.")
````

````python
print("\n--- Injeo por Atributo ---")
notifier = Notifier()

# Injetando a dependncia diretamente no atributo
sms_gateway = SMSGateway()
notifier.sender = sms_gateway
notifier.notify_user(user_id="user123", message="Seu pedido est a caminho!")

push_service = PushNotificationService()
notifier.sender = push_service
notifier.notify_user(user_id="user456", message="Nova promoo disponvel!")

notifier_2 = Notifier()
notifier_2.notify_user(user_id="user789", message="Erro do sender no definido.")
````

O LangChain  um framework em **Python (e JS)** criado para construo de aplicaes que usam LLMs (Large Language Models) como ChatGPT, Claude, Mistral, Llama, etc.

\

**Porque utilizar um Framework?**

* Ele facilita a criao de pipelines, chatbots, assistentes, agentes, RAGs (Retrieval-Augmented Generation), entre outros.

* Facilita a portabilidade entre as LLMs. (Cada LLM tem a sua API com as sua particularidade).

* Ele tem componentes prontos e bem separados (LLMs, Memory, Tools, Chains, Agents).

* Facilita a construo de pipelines complexas.

* J vem com recursos de tracking, observability, serialization, etc.

* Voc pode usar s as partes que quiser (no  obrigatrio usar tudo).

\

Para mais detalhes acesse: https://www.langchain.com/

**Groq** (https://groq.com)  uma empresa americana de inteligncia artificial fundada em 2016 por ex-engenheiros do Google. Seu principal diferencial e inovao reside no desenvolvimento de um circuito integrado especfico para aplicaes de IA que eles chamam de **LPU** (**Language Processing Unit**), e hardware relacionado.

A misso da Groq  acelerar o desempenho da inferncia de cargas de trabalho de IA, ou seja, o processo de usar um modelo de IA j treinado para gerar previses ou respostas. Eles se destacam por oferecer velocidade de processamento e eficincia incomparveis, superando as GPUs (Graphics Processing Units) nesse aspecto, que foram originalmente projetadas para processamento grfico e adaptadas para IA.

**Pontos Chave sobre o Groq:**

1. **LPU (Language Processing Unit)**:  o chip especializado da Groq. Diferente das GPUs, que so mais versteis, as LPUs foram projetadas especificamente para a inferncia de modelos de IA, especialmente Large Language Models (LLMs - Grandes Modelos de Linguagem). Essa especializao permite que as LPUs atinjam latncias ultrabaixas e alto throughput (taxa de gerao de tokens por segundo).
2. **Velocidade Instantnea**: A Groq tem ganhado destaque no mercado por sua capacidade de gerar respostas de LLMs quase instantaneamente. Eles frequentemente demonstram que seus sistemas podem gerar centenas ou at milhares de tokens por segundo, um desempenho significativamente mais rpido do que muitas outras solues disponveis.
3. **Eficincia Energtica e Custo**: Alm da velocidade, a arquitetura da LPU tambm  otimizada para maior eficincia energtica e menor custo por inferncia em comparao com as GPUs tradicionais.

\
Em resumo, a **Groq** se posiciona como uma alternativa poderosa  **NVIDIA** no espao de hardware de IA, focando especificamente em oferecer a inferncia de LLMs mais rpida e eficiente do mercado por meio de sua inovadora tecnologia LPU.

## Criando o Agente

````python
class AgenteDataFrame:

    # Vamos utilizar injeo de dependncia no construtor, para conseguimos,
    # trocar a LLM usada e o DataFrame sem ter que alterar nosso cdigo do
    # Agente. :)
    def __init__(self, llm:Type[BaseChatModel], df:pd.DataFrame) -> None:
        self.__df = df
        self.__llm = llm

    # No vamos criar do zero a ferramenta o LangChain tem vrias prontas! :)
    # https://python.langchain.com/docs/integrations/tools/
    @property
    def ferramentas(self) -> None:
        """
        Adiciona uma ferramenta ao agente.
        """
        return  [
                    Tool(
                        name="Cdigos Python",
                        func=PythonAstREPLTool(locals={"df": self.__df}),
                        description="""Utilize esta ferramenta sempre que o usurio solicitar clculos, consultas ou transformaes
                        especficas usando Python diretamente sobre o DataFrame `df`.
                        Exemplos de uso incluem: "Qual  a mdia da coluna X?", "Quais so os valores nicos da coluna Y?",
                        "Qual a correlao entre A e B?". Evite utilizar esta ferramenta para solicitaes mais amplas ou descritivas,
                        como informaes gerais sobre o DataFrame, resumos estatsticos completos ou gerao de grficos  nesses casos,
                        use as ferramentas apropriadas."""
                    )
                ]

    # Vamos criar o nosso prompt, baseado no ReAct (Reasoning + Acting)
    # O agente, a cada passo, faz:
    #   1. Pensamento (Thought / Reasoning): O LLM escreve o raciocnio do que est tentando fazer.
    #   2. Ao (Action): O LLM escolhe uma ferramenta e uma entrada (exemplo: chamar uma API, consultar um banco de dados,
    #      ler um PDF, fazer uma consulta SQL).
    #   3. Observao (Observation): O agente recebe a resposta da ferramenta (exemplo:
    #      "Resposta da API: 123 resultados encontrados").
    #   4. Volta ao passo 1 at chegar a uma Resposta Final (Final Answer).
    @property
    def react_prompt(self) -> str:
        """
        Define o prompt para o agente.
        """
        df_head:str = self.__df.head().to_markdown()
        return PromptTemplate(
                    input_variables=["input", "agent_scratchpad", "tools", "tool_names"],
                    partial_variables={"df_head": df_head},
                    template = """
                        Voc  um assistente que sempre responde em portugus.

                        Voc tem acesso a um dataframe pandas chamado `df`.
                        Aqui esto as primeiras linhas do DataFrame, obtidas com `df.head().to_markdown()`:

                        {df_head}

                        Responda s seguintes perguntas da melhor forma possvel.

                        Para isso, voc tem acesso s seguintes ferramentas:

                        {tools}

                        Use o seguinte formato:

                        Question: a pergunta de entrada que voc deve responder
                        Thought: voc deve sempre pensar no que fazer
                        Action: a ao a ser tomada, deve ser uma das [{tool_names}]
                        Action Input: a entrada para a ao
                        Observation: o resultado da ao
                        ... (este Thought/Action/Action Input/Observation pode se repetir N vezes)
                        Thought: Agora eu sei a resposta final
                        Final Answer: a resposta final para a pergunta de entrada original.

                        Comece!

                        Question: {input}
                        Thought: {agent_scratchpad}"""
                )

    def executar(self, pergunta:str) -> dict[str,str]:
        """
        Executa o agente com a entrada fornecida.
        """
        agente = create_react_agent(llm=self.__llm, tools=self.ferramentas, prompt=self.react_prompt)
        executor = AgentExecutor(agent=agente, tools=self.ferramentas, handle_parsing_errors=True)
        resposta = executor.invoke({"input": pergunta})
        return resposta
````

## Ferramentas de gerao de cdigo Python

**PythonREPLTool**
1. Executa cdigo Python diretamente
2. Usa `exec()` e `eval()`
3. Risco alto, pois o cdigo arbitrrio pode ser perigoso

\

**PythonAstREPLTool (Ast - Abstract Syntax Tree)**

1. Executa cdigo com parsing seguro via `ast`.
2. Usa anlise da rvore sinttica para restringir o cdigo permitido.
3. Mais seguro, pois impede execuo de cdigo perigoso

\

Para mais informaes pesquise sobre o **mdulo ast** do Python, [clique aqui](https://docs.python.org/3/library/ast.html)

## CSV utilizado no exemplo

Fonte: https://www.kaggle.com/datasets/benroshan/factors-affecting-campus-placement

\

Ele contm registros de estudantes de um campus (em um contexto **indiano/business school**), com foco em identificar fatores que influenciam se o estudante foi ou no colocado (empregado em estgio/emprego) no mercado de trabalho.

Inclui tanto aspectos acadmicos (notas no ensino mdio e superior) quanto de empregabilidade (especializao, experincia, salrio oferecido).


\

**Nome das colunas e descries**

**sl_no**: Este  um nmero de identificao nico para cada candidato no dataset. Pense nele como um ID de registro.

**gender**: Indica o gnero do candidato. 'M' representa Masculino e 'F' representa Feminino.

**ssc_p**: Refere-se  porcentagem de notas do Ensino Mdio (10 srie) do candidato.  a nota que o aluno tirou no que seria o equivalente ao fundamental II no Brasil, ou seja, 10 ano do ensino mdio (ensino mdio no Brasil tem 3 anos e nos EUA tem 4 anos).

**ssc_b**: Indica a instituio ou conselho educacional onde o candidato completou seu Ensino Mdio (10 srie). Pode ser "Central" (referindo-se a um conselho educacional centralizado, como o CBSE na ndia) ou "Others" (outros conselhos ou instituies).

**hsc_p**: Representa a porcentagem de notas do Ensino Superior (12 srie) do candidato.  a nota que o aluno tirou no que seria o equivalente ao ltimo ano do ensino mdio no Brasil.

**hsc_b**: Informa a instituio ou conselho educacional onde o candidato concluiu seu Ensino Superior (12 srie). Similar ao ssc_b, pode ser "Central" ou "Others".

**hsc_s**: Descreve a rea de especializao do candidato no Ensino Superior (12 srie). Por exemplo, "Science" (Cincias), "Commerce" (Comrcio) ou "Arts" (Artes).

**degree_p**: Corresponde  porcentagem de notas do candidato na sua graduao.

**degree_t**: Indica o tipo de graduao ou campo de estudo do candidato. Por exemplo, "Comm&Mgmt" (Comrcio e Gesto) ou "Sci&Tech" (Cincia e Tecnologia).

**workex**: Mostra se o candidato possui experincia de trabalho prvia. "Yes" (Sim) significa que ele tem experincia, e "No" (No) indica que no tem.

**etest_p**: Nota em teste de aptido (em percentagem)

**specialisation**: Especializao no MBA (Mkt&HR ou Mkt&Fin)

**mba_p**: Percentual no MBA

**status**: Se foi colocado ou no (Placed, Not Placed)

**salary**: Salrio mensal oferecido (zero ou vazio se no colocado)
